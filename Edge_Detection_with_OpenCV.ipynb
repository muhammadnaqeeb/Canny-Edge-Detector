> Firstly we grab an image 

> Then load it up in OpenCV

> This will give us representation of image as a numpy array

> Then we apply number of pre-processing steps before appaying final edge detection algorithm (Canny()) to find edges

* cv2.imread()

* cv2.cvtColor()

* cv2.GaussianBlur()

* cv2.Canny()



# **1. Import Dependencies**

```python
# Import opencv for computer vision stuff
import cv2



# Import matplotlib for visulizing an image
from matplotlib import pyplot as plt

from google.colab.patches import cv2_imshow
```

# **2. Load Image**

```python
# Read Image
img = cv2.imread('/content/mask.jfif')
#print(img)
# show image
cv2_imshow(img)
```

# **3. Edge Detection**




```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```python
# first step is to remove the noice in the image with a 5x5 Gaussian filter
blur = cv2.GaussianBlur(gray, (5,5),0)
cv2_imshow(blur)
```

```python
# Apply Canny
canny = cv2.Canny(blur, threshold1 = 200, threshold2 = 210)
```

```python
cv2_imshow(canny)
```

# **4. Export Edges**

```python
cv2.imwrite('edges.jpg',canny)
```
